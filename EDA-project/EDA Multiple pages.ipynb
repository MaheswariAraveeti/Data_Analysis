```python
Rank=[]
channel=[]
country=[]
Followers=[]
views=[]
Likes=[]
comments=[]
category=[]
for j in range(1,15):
    url=(f"https://hypeauditor.com/top-youtube/?p={j}")
    page=requests.get(url)
    
    soup=BeautifulSoup(page.text)
    
    for i in soup.find_all("div",class_="row-cell rank"):
        print(i.text)
        regex=re.findall(".*",i.text)
        #print(regex)
        if regex:
            Rank.append(regex[0])
        else:
            Rank.append(np.nan)
            
            
    for i in soup.find_all("div",class_="contributor__name-content"):
        print(i.text)
        regex=re.findall("^[a-zA-Z]+",i.text)
        print(regex)
        if regex:
            channel.append(regex[0])
        else:
            channel.append(np.nan)
            
            
    for i in soup.find_all("div",class_="row-cell audience"):
        print(i.text)
        regex=re.findall("\w+\s?\w+",i.text)
        print(regex)
        if regex:
            country.append(regex[0])
        else:
            country.append(np.nan)
            
    for i in soup.find_all("div",class_="row-cell subscribers"):
        print(i.text)
        regex=re.findall(".*",i.text)
        #print(regex)
        if regex:
            Followers.append(regex[0])
        else:
            Followers.append(np.nan)
            
    for i in soup.find_all("div",class_="row-cell avg-views"):
        print(i.text)
        regex=re.findall(".*",i.text)
        #print(regex)
        if regex:
            views.append(regex[0])
        else:
            views.append(np.nan)
            
            
    for i in soup.find_all("div",class_="row-cell avg-likes"):
        print(i.text)
        regex=re.findall(".*",i.text)
        #print(regex)
        if regex:
            Likes.append(regex[0])
        else:
            Likes.append(np.nan)
            
    for i in soup.find_all("div",class_="row-cell avg-comments"):
        print(i.text)
        regex=re.findall(".*",i.text)
        #print(regex)
        if regex:
            comments.append(regex[0])
        else:
            comments.append(np.nan)
            
    
    container=soup.find_all("div",class_="row__top")
    for i in container:
        a=i.find("div",class_="tag__content ellipsis")
        if a:
            category.append(a.text)
        else:
            category.append(np.nan)
    
        
    

    
    
        
        
    
```

```python
print(len(Rank))
print(len(channel))
print(len(country))
print(len(Followers))
print(len(views))
print(len(Likes))
print(len(comments))
print(len(category))
```
